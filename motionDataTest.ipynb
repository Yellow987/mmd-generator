```python
import pymeshio.vmd.reader
import pandas as pd
import numpy as np
```

```python
motionData = pymeshio.vmd.reader.read_from_file('data/PV311_LIVE_MIK_00.vmd')
print(motionData)
```

```python
boneFrames = [{'frame': motion.frame, 'name': motion.name.decode("shift-jis"), 'position': motion.pos, 'rotation': motion.q, 'complement': motion.complement} for motion in motionData.motions]
df = pd.DataFrame(boneFrames)
print(df)
```

```python
count_series = df['name'].value_counts()

print(count_series.tail(40))
```

```python
filtered_df = df[df['name'] == 'センター']

print(filtered_df)
```
